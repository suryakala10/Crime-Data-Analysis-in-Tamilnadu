```python
 df['Violent_Crimes'] = df[['Rape', 'Attempt to Commit Rape', 'Acid Attack', 'Attempt to Acid Attack',
                           'Assault on Outrage Modesty of Women', 'Insult to the Modesty of Women',
                           'Abetment to Suicide of women', 'Dowry Deaths']].sum(axis=1)

df['Sexual_Exploitation'] = df[['POCSO', 'Buying of Minor Girls', 'Selling of Minor Girls',
                                'Human Trafficking', 'Immoral Traffic (Prevention) Act']].sum(axis=1)

df['Domestic_Crimes'] = df[['Cruelty by Husband or his relatives', 'Dowry Prohibition Act',
                            'Protection of Women from Domestic Violence Act']].sum(axis=1)

df['Cyber_Offenses'] = df[['Cyber Crimes/Information Technology Act', 
                           'Indecent Representation of Women (Prohibition) Act']].sum(axis=1)

df['Kidnapping_Abduction'] = df[['Kidnapping & Abduction']]
```

```python
crime_categories = {
    "Violent Crimes": df['Violent_Crimes'],
    "Sexual Exploitation": df['Sexual_Exploitation'],
    "Domestic Crimes": df['Domestic_Crimes'],
    "Cyber Offenses": df['Cyber_Offenses'],
    "Kidnapping & Abduction": df['Kidnapping_Abduction']
}

df_grouped = pd.DataFrame(crime_categories)
```

```python
# Setting 'Year' as Index
df_grouped['Year'] = df['Year'] 
df_grouped.set_index('Year', inplace=True)
```

## Observation ##
- The `Year` column was successfully added to the `df_grouped` DataFrame and then set as its

```python
print(df_grouped.head())
```

```python
df_grouped.plot(figsize=(12, 6), title='Category-wise Crime Trends in Tamil Nadu (2014Present)')
plt.ylabel('Number of Cases')
plt.xlabel('Year')
plt.grid(True)
plt.show()
```

## Observation ##
- Category-wise Crime Trends
- The line plot visualizes crime trends across five major categories 

```python
sns.heatmap(df_grouped.corr(), annot=True, cmap='coolwarm')
```

## Observation ##
-Correlation Heatmap of Crime Categories
- The heatmap displays the **correlation coefficients** between the major crime categories in Tamil Nadu .

```python
category_totals = df_grouped.sum().sort_values(ascending=False)

sns.barplot(x=category_totals.values, y=category_totals.index, palette='Reds_r')
plt.title('Total Crimes by Category ')
plt.xlabel('Total Crime Count')
plt.ylabel('Crime Category')
plt.tight_layout()
plt.show()
```

## Observationn ##
- The bar plot clearly shows that **Violent Crimes** have the **highest total crime count** over the 10-year period.


```python
category_totals = df_grouped.sum()

plt.figure(figsize=(8, 8))
plt.pie(category_totals, labels=category_totals.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Reds_r', len(category_totals)))
plt.title('Proportion of Total Crimes by Category')
plt.axis('equal') 
plt.tight_layout()
plt.show()
```

## Observation ##
-The pie chart shows the **relative share of each crime category** over the 10-year period.

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_grouped, x='Violent Crimes', y='Domestic Crimes', s=100, color='darkred')

plt.title('Violent Crimes vs Domestic Crimes (Tamil Nadu)')
plt.xlabel('Violent Crimes')
plt.ylabel('Domestic Crimes')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Observation ##
- The scatter plot shows a **strong positive relationship** between **Violent Crimes** and **Domestic Crimes** over the years.
